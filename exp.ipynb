```python
import matplotlib.pyplot as plt
import numpy as np

from optimization.algorithms import (
    Parameter,
    adaptive_regularized_newton,
    gradient_descent,
    newton,
    trust_region,
)
from optimization.problem import generate_Rosenbrock
```

```python
gen_prb = generate_Rosenbrock
param = Parameter()

algs = [
    gradient_descent,
    newton,
    trust_region,
    adaptive_regularized_newton,
]
alg_names = [
    "Gradient Descent",
    "Newton",
    "Trust-Region",
    "ARN",
]

n = 10
mu = -1
seed = 42
```

```python
problem = gen_prb(n=n, mu=mu, seed=seed)

fig = plt.figure(figsize=(10,4))

ax_f = fig.add_subplot(1, 2, 1)
ax_g = fig.add_subplot(1, 2, 2)

for alg, label in zip(algs, alg_names):
    sols, y, grad = alg(problem, param)
    ax_f.plot(range(len(y)), y, label=label)
    ax_g.plot(range(len(grad)), grad, label=label)

ax_f.set_ylabel("$f(x_k)$")
ax_g.set_ylabel("||grad $f(x_k)$||")
for ax in (ax_f, ax_g):
    ax.set_yscale("log")
    ax.set_xlim([0, 100])
    ax.set_xlabel("$k$")
    ax.legend()
plt.show()
```
